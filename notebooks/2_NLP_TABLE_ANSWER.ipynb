# Table Question Answering using Google's tapas

```python
from transformers import pipeline
import pandas as pd

tqa = pipeline(task="table-question-answering", model="google/tapas-large-finetuned-wtq")
```

```python
# prepare table
data = {
    "Package": [
        "Char Dham Yatra", "Singapore/Malaysia", "Pink City - Jaipur", "Europe Tour"
    ],
    "Category": [
        "Domestic", "International", "Domestic", "International"
    ],
    "Duration": [
        "5 Days/4 Nights", "7 Days/6 Nights", "3 Days/2 Nights", "15 Days/14 Nights"
    ],
    "Cost INR": [
        "30000", "75000", "25000", "200000"
    ]
}
table = pd.DataFrame.from_dict(data)
table
```

```python
# lets ask some questions | inherently it takes care of SQL execution
question = "Which tours are domestic?"
print(tqa(table=table, query=question)['cells'])
```

```python
#some more questions
question = "How long is the Europe Tour?"
print(tqa(table=table, query=question)['cells'])

question = "Which category does the Jaipur belong to?"
print(tqa(table=table, query=question)['cells'])
```
